## Algorithms

Within computer science, the term algorithm is used quite generally. It refers to a step-by-step procedure to solve a problem and does not have to be written in a programming language.

One example of an algorithm in everday life is making a cup of tea:

1. Boil Water
2. Place tea bag in a mug
3. Pour hot water into mug
4. Let the tea bag steep
5. Remove tea bag
6. (Optional) Add milk or sugar

This example follows the constructs of an algorithm as it has a clear goal, it follows a sequence of steps, and it can be automated.

After chatting with my peer group a few more examples we came up with were:
1. Following a recipe
2. Going to the shops and using a list
3. Skincare routine/Makeup routine

## Pseudocodes

Pseudocodes differ from algorithms because they adopt basic programming concepts where algorithms do not to make it 'code-like'. However, similar to algorithms it does not have to be written in a coding language and can still be in English.

Using the previous example, I converted it into an example of a pseudocode:

BOIL water

PLACE tea bag in mug

POUR water into mug

WAIT for tea to steep

REMOVE tea bag

IF user wants milk THEN

 ADD milk

ENDIF

If user wants sugar THEN

 ADD sugar

ENDIF

After discussing with my peer group, some differences we noticed are that pseudocodes are much more succinct. It simplifies the instructions down to the most basic tasks. Also, although it is written in English, it follows coding languages where the task is in bold.

Another example a peer gave of pseudocode is of a simple addition sum:

DEF 

create variable that holds values a and b

insert addition function a + b

PRINT result of addition

INSERT values to be added

OUTPUT result


This example more closely alligns with the language of Python, using functions such as PRINT and DEF.

## Python Code

The following is an example of a code that analyses the most common word in a poem. Before running the code, my guess that the most common word will be both 'more' and 'summers':

```python
from collections import Counter 

poem = "Shall I compare thee to a summer's day? Thou art more lovely and more temperate: Rough winds do shake the darling buds of May, And summer’s lease hath all too short a date:" 

words = [w.lower() for w in poem.split()] 

counts = Counter(words) 

most_common = counts.most_common(1)[0] 

print ("Most frequent word:", most_common[0], "appears", most_common[1], "times") 
```

Unlike my predicition and ouput 'a' as the most common word. I spoke with my peer group to assess the code. We agreed that it is not fully effective as there a multiple words that all have appear twice, and the code only allows one to be displayed. I then consulted with MS Copilot to trouble shoot the code. The new output suggested by Copilot was:

```python
from collections import Counter

poem = "Shall I compare thee to a summer's day? Thou art more lovely and more temperate: Rough winds do shake the darling buds of May, And summer’s lease hath all too short a date:"

# Normalize and clean words
words = [w.lower().strip(".,:;!?") for w in poem.split()]

# Count word frequencies
counts = Counter(words)

# Show top 5 most common words
top_n = 5
most_common_words = counts.most_common(top_n)

# Print results
print(f"Top {top_n} most frequent words:")
for word, freq in most_common_words:
    print(f"'{word}' appears {freq} times")
```

After discussing the output with my peer group, we agreed that Copilot in this case over-corrected the error. It changed the code to include the top 5 words, however, this means that the code is still negating words that should be included in the output. We came to the conclusion that the prompt was not specific enough for Copilot to complete the task perfectly. However, this does show that the user must be very specific in order to achieve the correct output from Copilot.

I then asked Copilot to fix its own mistake to get the intended output:

```python
from collections import Counter

poem = "Shall I compare thee to a summer's day? Thou art more lovely and more temperate: Rough winds do shake the darling buds of May, And summer’s lease hath all too short a date:"

# Normalize and clean words
words = [w.lower().strip(".,:;!?") for w in poem.split()]

# Count word frequencies
counts = Counter(words)

# Find the highest frequency
max_freq = max(counts.values())

# Get all words with that frequency
most_common_words = [word for word, freq in counts.items() if freq == max_freq]

# Print results
print(f"Most frequent word(s) appearing {max_freq} time(s):")
for word in most_common_words:
    print(f"'{word}'")
```

In this case it did fix the error to achieve the correct output and was able to resolve the issue. This suggests that althout Copilot is a useful tool, it is not always correct and mutliple attempts may be needed to achieve the needed output.
