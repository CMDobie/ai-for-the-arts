## Python Libraries

Python libraries can be imported to provide access to pre-written could that serve functional tasks.

Remembering back to the Part 1 notebook, some examples of libraries are:
1. math - Built-in library for basic math functions
2. numpy - Handles large arrays and matrices
3. scipy - Builds on numpy for more advanced math

In this notebook an important libray is IPython.display. This library has command for displaying videos. In order to install it the correct syntax is:

*pip install IPython.display*

'pip' being a package manager that installs the selected library.

Once the library has been installed, it can then be used to provide more functionality in code. To do so, at the beginning of code the key word 'from' is used to import whatever command is needed. An example is given below:

```python
from IPython.display import *
```

In this example an * is used to import all commands from the library.
