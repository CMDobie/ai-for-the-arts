#### 1. if, elif, else Statements

The first prompt I entered was in relation to if, elif, and else statements:

*"Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concepts of if, elif, and else statements to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

The output was:

```python
greeting = "Hello"

if greeting == "Hi":
    print("Hi there!")
elif greeting == "Hello":
    print("Hello, world!")
else:
    print("Greetings, traveler!")
```

From this output it can be understood that 'if' is a method of fact checking a list of conditions. Using variables the code cross-references the originally stated variable against the provided list of potential 'if' statements to output the associated greeting. I then attempted to replicate the code with the output being the greeting held under 'else':

```python
greeting = "Good morning"

if greeting == "Hi":
    print("Hi there!")
elif greeting == "Hello":
    print("Hello, world!")
else:
    print("Greetings, traveler!")
```

#### 6. Importing Libraries

This prompt was entered into Copilot to understand how to import libraries in Python:

*"Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concept of importing libraries to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

The following code was produced from this prompt:

```python
import random

greetings = ["Hello", "Hi", "Hey", "Greetings", "Salutations"]

selected_greeting = random.choice(greetings)

print(selected_greeting + ", world!")
```

The import function is used to import whatever tool is written next to it, in this case 'random'. This is a library that allows the selection of random values. Once again a list is created held under 'greetings'. The randomchoice() is a function from the random library that will pick one value held within the list, which in this case is stored in the variable 'selected_greeting'.

#### 8. Classes, Objects, and Methods

The following code was entered to Copilot:

*"Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concept of classes, objects and methods to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

Copilot generated the following code:

```python
class Greeter:
    def say_hello(self):
        print("Hello, world!")

my_greeter = Greeter()
my_greeter.say_hello()
```

The first line of code defines the class. The class is a keyword that defines a blueprint for creating objects; Greeter is the class, with the following code explaining what Python to do with it.

A method is a function that belongs to a class. The next two lines of code define the method; say_hello is the name of the method which prints "Hello, World" whenever called.

The fourth line creates an object; my_greeter is a variable that holds the object of greeter.

Finally the method is called using the object and method.
