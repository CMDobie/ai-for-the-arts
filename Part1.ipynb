#### 1. if, elif, else Statements

The first prompt I entered was in relation to if, elif, and else statements:

*"Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concepts of if, elif, and else statements to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

The output was:

```python
greeting = "Hello"

if greeting == "Hi":
    print("Hi there!")
elif greeting == "Hello":
    print("Hello, world!")
else:
    print("Greetings, traveler!")
```

From this output it can be understood that 'if' is a method of fact checking a list of conditions. Using variables the code cross-references the originally stated variable against the provided list of potential 'if' statements to output the associated greeting. I then attempted to replicate the code with the output being the greeting held under 'else':

```python
greeting = "Good morning"

if greeting == "Hi":
    print("Hi there!")
elif greeting == "Hello":
    print("Hello, world!")
else:
    print("Greetings, traveler!")
```

After chatting with my peer group it should be noted that a different example was given as a response to a peer:

```python
time_of_day = "morning"

if time_of_day == "morning":
    print("Good morning! Hello, world!")
elif time_of_day == "afternoon":
    print("Good afternoon! Hello, world!")
else:
    print("Good evening! Hello, world!")
```

Even though both make use of the conditioning of the 'if' code, and output a very similar result, both examples have different variables and conditions.  It is interesting that where before, with simple prompts, the output was the same across the peer group, as the prompt becomes more complex Copilot has more variation in response. From this it can be understood that although for these simple purposes Copilot remains accurate, it does output the same ideas based off of the same prompt.

From the response it can be understood that loops are a block of code that are repeated a number of times. The example given by Copilot was:

#### 3. Data Structures/Lists

The initial prompt given to Copilot about data structures was:

*"Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concept of lists to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

```python
greetings = ["Hello, world!", "Hi there!", "Greetings, Earthling!", "Hey, everyone!", "Salutations!"]

for message in greetings:
    print(message)
```
