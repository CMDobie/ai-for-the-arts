## Using MS Copilot

In order to better understand the use of MS Copilot I submitted the following prompt into Copilot:

*"Could you suggest a python code for ‘Hello, world!’ to demonstrate the concept of printing to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects."*

The code produced was:

```python
print("Hello, World!")
```

To better understand the uses of Copilot I deliberately typed the wrong code and asked Copilot to trouble-shoot the issue with the following prompt:

*I submitted the following code in Python and received an error:* print(Hello, World!) . *Suggest what is wrong with this and what I can do to fix the error.*

In reponse MS Copilot spotted that I had missed out the quotation marks surrounding my intended output "Hello, World!". Once I replaced them, as suggested by Copilot, the code continued to work normally.

From this initial introduction to Copilot, it does seem like an effective tool for writing code and also for spotting errors.
